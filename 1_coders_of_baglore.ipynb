```python
with open("initialdata.txt", encoding="utf") as f:
    data = f.read()
```

```python
chunks = data.split('\n\n')
```

```python
chunks = [c for c in chunks if len(c)>3]
```

```python
print(chunks[1])
```

```python
print(chunks[0])
```

```python
def extracting_data(chunk):
    chunk = chunk.strip()
    sparc_chunk = chunk.split('\n')
    user_name = sparc_chunk[0]
    post = sparc_chunk[1]
    followers = sparc_chunk[2]
    following = sparc_chunk[3]
    name = sparc_chunk[4]
    if (len(sparc_chunk) > 5) :
        org = sparc_chunk[5]
        bio = sparc_chunk[6:]
        # print(org)
        
    else :
        org = "Unknown"
        bio  = ""
        # print(org)
    return { "user_name" : user_name  , "post" : post , "followers" : followers , "following" : following , "name" : name , 
            "org" : org , "bio" : bio }
    
```

```python
parsed_data = []
for chunk in chunks: 
    parsed_chunked = extracting_data(chunk)
    parsed_data.append(parsed_chunked)
    
```
