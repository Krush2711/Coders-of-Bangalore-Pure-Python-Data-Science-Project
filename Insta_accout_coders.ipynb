# Coders of Bangalore
You ask Sam Altman for a 2.2 billion USD in funding and he makes fun of you in public

He asks you where you live and finally after talking to you, gives you a challenging task 

Collect raw instagram data of all openai followers

Answer these questions:

- Who has maximum posts
- Who has maximum followers 
- Who follows maximum people 
- How many categories (Digital creators, Non profit foundation etc do we have) of how many people do we have?

You have 24 hours 

You then ask for data - he laughs!

```python
with open("finaldata.txt", encoding="utf") as f:
    data = f.read()
```

```python
chunks = data.split('\n\n')
```

```python
chunks = [c for c in chunks if len(c)>3]
```

```python
def extracting_data(chunk):
    chunk = chunk.strip()
    sparc_chunk = chunk.split('\n')
    user_name = sparc_chunk[0]
    post = int(sparc_chunk[1].split(' post')[0].replace(',', "").replace('K', "000").replace("M", "000000" ), )
    followers = float(sparc_chunk[2].split(' follower')[0].replace(',', "").replace('K', "").replace("M", "" ), )
    if "K" in sparc_chunk[2]:
        followers = int(followers*1000)
    elif "M" in  sparc_chunk[2]:
        followers = int(followers*1000000)
    else:
        followers = int(followers)
    
    following = float(sparc_chunk[3].split(' following')[0].replace(',', "").replace('K', "").replace("M", "" ), )
    if "K" in sparc_chunk[3]:
        following = int(following*1000)
    elif "M" in  sparc_chunk[3]:
        following = int(following*1000000)
    else:
        following = int(following)
    
    name = sparc_chunk[4]
    if (len(sparc_chunk) > 5) :
        org = sparc_chunk[5]
        bio = sparc_chunk[6:]
        # print(org)
        
    else :
        org = "Unknown"
        bio  = ""
        # print(org)
    return { "user_name" : user_name  , "post" : post , "followers" : followers , "following" : following , "name" : name , 
            "org" : org , "bio" : bio }
    
```

```python
parsed_data = []
for chunk in chunks: 
    parsed_chunked = extracting_data(chunk)
    parsed_data.append(parsed_chunked)
    
```

```python
import json 
js = json.dumps(parsed_data , indent=4)
with open("data.json" , "w") as f :
    f.write(js)
```

```python
max = 0
for chunk in parsed_data:
    if (max < chunk['post']):
        max = chunk['post']
        max_posts = chunk
print(max_posts)
    
```
